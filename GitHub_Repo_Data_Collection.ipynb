How to use the Python programming language to collect data from the GitHub API

- Parse through the response to save the data into variables. 
- Using a for loop to make a second API call to grab repo statistics. 

```python
import requests #make API calls
import pandas as pd
import time
```

```python
# GitHub API endpoint to fetch repositories of a user
USERNAME = "saranobrega"  # Example GitHub username
API_URL =  f"https://api.github.com/users/{USERNAME}/repos"

# Make API call
response = requests.get(API_URL)

# Check if the request was successful
if response.status_code == 200:
    repositories = response.json()
    for repo in repositories:
        print(repo['name'])  # Print the name of each repository
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

# Creating a Loop

```python
for repo in response:
    name = repo['name']
    description = repo.get('description', 'No description')  # Some repos might not have a description
    created_at = repo['created_at'].split('T')[0]  # Extracting the date part before 'T'
    
    print(f"Name: {name}")
    print(f"Description: {description}")
    print(f"Created At: {created_at}")
    print("----------")
```

```python
# Create an empty list to store the repository data
data = []

for repo in response:
    name = repo['name']
    description = repo.get('description', 'No description')  # Some repos might not have a description
    created_at = repo['created_at'].split('T')[0]  # Extracting the date part before 'T'
    
    # Append the data to the list as a dictionary
    data.append({
        'Repo_Name': name,
        'Description': description,
        'Created_At': created_at
    })

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(data)
```
